```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
nn = 250
rand = nx.erdos_renyi_graph(nn,0.01)

plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.title("Erdos-Renyi")
nx.draw(rand,node_size=10)
# plt.show()

plt.subplot(1, 3, 2)
degree_sequence = sorted([d for n, d in rand.degree()], reverse=True)
plt.hist(degree_sequence, bins=20)
plt.title("Degree Distribution")
plt.xlabel("Degree")
plt.ylabel("Frequency")
# plt.show()


graph = rand
clus_cof = []
for i in range(nn):
  clus_cof.append(nx.clustering(graph,i))
# print(clus_cof)


plt.subplot(1, 3, 3)
plt.hist(clus_cof)
plt.title("Clustering coefficient histogram ")
plt.tight_layout()

# Display mean degree, average shortest path, and max clustering by node
mean_degree = sum(degree_sequence) / len(degree_sequence)
# avg_shortest_path = nx.average_shortest_path_length(graph)
max_clustering_node = np.argmax(clus_cof)
max_clustering_value = max(clus_cof)

mean_degree = sum(degree_sequence)/len(degree_sequence)
print("Mean value of degree of nodes is ", mean_degree)
# print("Average shortest path =",nx.average_shortest_path_length(graph))
print("Max clustering by node", max_clustering_node, "is", max_clustering_value)

plt.show()
```

```python
# average shortest path of a random networkimport networkx as nx
import matplotlib.pyplot as plt
import numpy as np

nn = 250
sw = nx.watts_strogatz_graph(nn, 4, 0.1)
graph = sw

plt.figure(figsize=(15, 5))

# Plot 1: Small-World
plt.subplot(1, 3, 1)
plt.title("Small-World")
nx.draw(graph, node_size=50, with_labels=False)

# Plot 2: Degree Distribution
plt.subplot(1, 3, 2)
degree_sequence = sorted([d for n, d in graph.degree()], reverse=True)
hist, bins, _ = plt.hist(degree_sequence, bins=20)
plt.title("Degree Distribution")
plt.xlabel("Degree")
plt.ylabel("Frequency")


# Plot 3: Clustering Coefficient Histogram

clus_cof = [nx.clustering(graph, i) for i in range(nn)]
plt.subplot(1, 3, 3)
plt.hist(clus_cof)
plt.title("Clustering Coefficient Histogram ")
plt.tight_layout()


# Display mean degree, average shortest path, and max clustering by node
mean_degree = sum(degree_sequence) / len(degree_sequence)
avg_shortest_path = nx.average_shortest_path_length(graph)
max_clustering_node = np.argmax(clus_cof)
max_clustering_value = max(clus_cof)

print("Mean value of degree of nodes is ", mean_degree)
print("Average shortest path =", avg_shortest_path)
print("Max clustering by node", max_clustering_node, "is", max_clustering_value)

plt.show()
```

```python
nn = 125
init_graph = nx.scale_free_graph(nn, alpha=0.9, beta=0.05, gamma=0.05)
init_graph = nx.Graph(init_graph) #Converting multigraph to simple graph
graph = init_graph

plt.figure(figsize=(15,5))
plt.subplot(1, 3, 1)
plt.title("Initial Scale-free Network (N =" + str(nn)+")")
nx.draw(graph, node_size=10, node_color = 'red')

degree_sequence = sorted([d for n, d in graph.degree()], reverse=True)
plt.subplot(1, 3, 2)
plt.hist(degree_sequence, bins=20)
plt.title("Degree Distribution")
plt.xlabel("Degree")
plt.ylabel("Frequency")


clus_cof = []
for i in range(nn):
    clus_cof.append(nx.clustering(graph, i))

plt.subplot(1, 3, 3)
plt.hist(clus_cof)
plt.title("Clustering coefficient histogram ")
plt.tight_layout()


mean_degree = sum(degree_sequence) / len(degree_sequence)
print("Mean value of degree of nodes is ", mean_degree)
print("Average shortest path =",nx.average_shortest_path_length(graph))
print("Max clustering by node ",np.argmax(clus_cof),"is",max(clus_cof))

plt.show()
```
